```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.columns = ['gender',
              'race_ethnicity',
              'parental_level_of_education',
              'lunch',
              'test_preparation_course',
              'math_score',
              'reading_score',
              'writing_score'
             ]
df
```

```python
df = df[np.abs(df['math_score']-df['math_score'].mean()) <= (3*df['math_score'].std())]
df = df[np.abs(df['reading_score']-df['reading_score'].mean()) <= (3*df['reading_score'].std())]
df_cleaned = df[np.abs(df['writing_score']-df['writing_score'].mean()) <= (3*df['writing_score'].std())]
df_cleaned
```

```python
# quantify parental_level_of_education
df_cleaned['parental_level_of_education'] = df_cleaned['parental_level_of_education'].map({'some high school': 0,
                                                                                          'high school': 1,
                                                                                          "associate's degree": 2,
                                                                                          'some college': 3,
                                                                                          "bachelor's degree": 4,
                                                                                          "master's degree": 5})
```

```python
# quantify race_ethnicity
dummies1 = pd.get_dummies(df_cleaned.race_ethnicity)
dummies1
```

```python
# quantify gender
dummies2 = pd.get_dummies(df_cleaned['gender'])
dummies2
```

```python
#quantify lunch
df_cleaned['lunch'] = df_cleaned['lunch'].map({'free/reduced': 0,'standard': 1})
```

```python
# merge dummies into df
df_quant = pd.concat([df_cleaned,dummies1,dummies2],axis='columns')
df_quant
```

```python
# drop original gender and race_ethnicity
# drop group A and male to avoid dummy trap
df_quant = df_quant.drop(['gender','race_ethnicity','group A','male'],axis='columns')
df_quant
```

```python
import warnings
warnings.filterwarnings('ignore')
from pymatch.Matcher import Matcher
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
# create test and control groups by test_preparation_course
test = df_cleaned[df_quant.test_preparation_course == "completed"]
control = df_cleaned[df_quant.test_preparation_course == "none"]
test['test_preparation_course'] = 1
control['test_preparation_course'] = 0
```

```python
# Matcher
m = Matcher(test, control, yvar="test_preparation_course", exclude=['math_score','reading_score','writing_score'])
```

```python
# for reproducibility
np.random.seed(20170925)
m.fit_scores(balance=True, nmodels=358)
```

```python
# pridict scores
m.predict_scores()
m.plot_scores()
```

```python
# tune threshold
m.tune_threshold(method='random')
```

```python
# match data
m.match(method="min", nmatches=1, threshold=0.0007)
m.record_frequency()
```

```python
m.assign_weight_vector()
pairs = m.matched_data.sort_values("match_id")
pairs['total_score'] = pairs['math_score'] + pairs['reading_score'] + pairs['writing_score']
course = pairs.loc[pairs['test_preparation_course'] == 1]
none = pairs.loc[pairs['test_preparation_course'] == 0]
none
```

```python
# Calculate mean and std of total_scores of with_lunch group and without_lunch group 
miu1 = np.mean(course['total_score'])
miu2 = np.mean(none['total_score'])
sigma1 = np.std(course['total_score'])
sigma2 = np.std(none['total_score'])
sigma1_2 = np.sqrt((sigma1**2)/358 + (sigma2**2)/358)
print(sigma1_2, miu1-miu2)
```
